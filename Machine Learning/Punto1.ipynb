### Se importan las bibliotecas necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
import matplotlib.animation as mpla
import math
```

### Se crea el vector de puntos X y la función
$$ y = \sqrt{x} $$

```python
X = np.linspace(0,50, 800).reshape(-1,1)
y = np.sqrt(X).ravel()
plt.plot(X,y)
plt.xlabel("x")
plt.ylabel("Valores reales de y")
plt.title("Función sqrt(x)")
plt.show()
```

## Entrenamiento del modelo.
Ya que la función que se va a predecir no es lineal, se escogió como modelo un Suport Vector Machine. Los parámetros que se le dan son los siguientes:
- kernel: la función que se usa para transformar los datos es gaussiana.
- C: Es un parámetro regularizador. Controla que los parámetros w no se sobreajusten.
- epsilon: Es el margen de tolerancia para el error de predicción.

```python
model = SVR(kernel='rbf', C=100, gamma=0.01, epsilon=0.1)
model.fit(X,y.ravel())
```

```python
X_test = np.linspace(0, 50, 10).reshape(-1,1)
y_test = np.sqrt(X_test)
y_pred = model.predict(X_test)

plt.plot(X, y, label="Función real")
plt.plot(X_test, y_pred, label="Predicción de y")
plt.plot(X_test, y_test, label="Validación de y")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Predicción del modelo")
plt.legend()
plt.show()
```
